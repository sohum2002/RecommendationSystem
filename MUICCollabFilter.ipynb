#Project

This project is a Recommendation System using Collaboration Filtering. I am predicting how much a person would enjoy a class (out of 5) based on their previous ratings of other classes and their general preferences (How much they enjoy, for example, Homework, Projects etc). The dataset is collected by myself through survey-based questionnaires.

```python
with open('classRatingReal.csv') as f:
    s = StringIO.StringIO(f.read()) 
    reader = csv.reader(s, delimiter=',')
    raw = [row for row in reader]

with open('personRatingReal.csv') as f:
    s2 = StringIO.StringIO(f.read())
    reader2 = csv.reader(s2, delimiter=',')
    raw2 = [row for row in reader2]
    
def get_names(raw):
    return raw[0][1:]

def get_classes(raw):
    classes = []
    for i in range(1, len(raw)):
        classes.append(raw[i][0])
    return classes

def get_feature_names(raw2):
    feature_names = []
    for i in range(1, len(raw2)):
        feature_names.append(raw2[i][0])
    return feature_names

def get_rating(raw):

    names = get_names(raw)
    classes = get_classes(raw)
    n_names = len(names)
    n_rests = len(classes)

    rests_offset = 1
    names_offset = 1

    has_rating = np.zeros((n_names, n_rests))
    rating = np.zeros((n_names, n_rests))
    
    for ir in range(n_rests):
        this_row = raw[ir+rests_offset]
        for ip in range(n_names):
            this_cell =  this_row[ip + names_offset]
            
            hr = this_cell.isdigit()
            
            has_rating[ip, ir] = 1 if hr else 0
            if hr:
                rating[ip, ir] = int(this_cell)

    return names, classes, has_rating, rating

def get_feature(raw2):
    names = get_names(raw2)
    feature_names = get_feature_names(raw2)
    
    n_names = len(names)
    n_features = len(feature_names)
    
    features_offset = 1
    names_offset = 1
    
    feature_rating = np.zeros((n_names, n_features))
    
    for i in range(n_features):
        this_row = raw2[i+features_offset]
        for j in range(n_names):
            this_cell = this_row[j + names_offset]
            hr = this_cell.isdigit()
            
            feature_rating[j,i] = int(this_cell)
    
    return names, feature_names, feature_rating
            

names, classes, has_rating, rating = get_rating(raw)

get_feature(raw2)

namesFromFeatures, feature_names, feature_rating = get_feature(raw2)

plt.matshow(rating, vmin=0, vmax=5)

plt.matshow(feature_rating, vmin=0, vmax=5)

assert((len(names) == len(namesFromFeatures))), "Names dimensions don't match"
assert((len(rating[0].shape) == len(feature_rating[0].shape))), "Rating dimensions don't match"
```

```python
class collabFilter:
    def __init__(self,names, classes, has_rating, rating, namesFromFeatures, feature_names, feature_rating):
        self.names = names
        self.classes = classes
        self.has_rating = has_rating
        self.rating = rating
        self.namesFromFeatures = namesFromFeatures
        self.feature_names = feature_names
        self.feature_rating = feature_rating
        
    def compute(self):
        f = np.random.rand(133,7)
        p = self.feature_rating
        
        for i in xrange(100000):
            hgr_out = self.hgr(f,p)
            dp = self.dp(hgr_out,f)
            df = self.df(hgr_out,p)
            
            p = p - (0.001 * dp)
            f = f - (0.001 * df)
        guess_rating = np.dot(p,np.transpose(f))
        return guess_rating
            
    def hgr(self,f,p):
        return has_rating * (np.dot(p,np.transpose(f)) - rating)
        
    def dp(self,cost1,f):
        return np.dot(cost1,f)

    def df(self,cost1,p):
        return np.dot(np.transpose(cost1),p)
    
cf = collabFilter(names, classes, has_rating, rating, namesFromFeatures, feature_names, feature_rating)

G = cf.compute()
```

## The next bit is written to visualize the results in a HTML - table format

```python
from IPython.display import HTML

class TableCell:
    
    def __init__(self, text, tc=None, color=None):
        self.text = text
        self.tc = tc
        self.color = color
    
    def to_html(self):
        return '<td>%s</td">'%self.text
    
def maketable(rating, has_rating, guess, classes, names):
    n_rests = len(classes)
    n_names = len(names)
    tab = np.empty((n_rests+1, n_names+1),dtype='object')
    #print tab.shape

    for irest in range(n_rests):
        tab[irest+1,0] = classes[irest]

    for iname in range(n_names):
        tab[0,iname+1] = names[iname]

    for irest in range(n_rests):
        for iname in range(n_names):
            if not has_rating[iname, irest]:
                tab[irest+1, iname+1] = TableCell('<span style="color:red">%3.2f</span>'%(guess[iname, irest]))
            else:
                tab[irest+1, iname+1] = TableCell('<span style="color:blue">%3.2f</span><span style="color:red">(%3.2f)</span>'%(rating[iname, irest], guess[iname, irest]))
    #now convert tab array to nice html table
    nrow, ncol = tab.shape
    t = []
    t.append('<table>')
    for irow in range(nrow):
        t.append('<tr>')
        for icol in range(ncol):
            cell = tab[irow,icol]
            if cell is not None:
                if isinstance(cell,TableCell):
                    t.append(tab[irow, icol].to_html())
                else:
                    t.append('<td>')
                    t.append(tab[irow, icol])
                    t.append('</td>')
            else:
                t.append('<td></td>')
        t.append('</tr>')
    t.append('</table>')

    return '\n'.join(t)
```

```python
# call it like this

HTML(maketable(rating, has_rating, G, classes, names))
```
